### Normalizar as variáveis 'preço' e 'mínimo de noites' usando z-score

- z-score vai normalizar os dados pela média (z-score = 0) e desvio padrão (para 1 desvio padrão, z-score = -1 ou +1).

 


```python
z = np.abs(stats.zscore(Dados_treino['price']))
Dados_treino['price_z'] = z

z_noite = np.abs(stats.zscore(Dados_treino['minimum_nights']))
Dados_treino['night_z'] = z_noite


plt.subplots(nrows=2, ncols=1, figsize=(100, 10))

plt.subplot(2, 1, 1)
sns.boxplot(x=Dados_treino['price_z'])

plt.subplot(2, 1, 2)


sns.boxplot(x=Dados_treino['night_z'])
```

- É possível identificar que existem valores que estão a mais de 20, 30 e até 100 desvios padrões da média e que influenciam negativamente a análises de regressão e modelos preditivos, pois possuem um peso muito grande. 

- Por convenção os outliers geralmente são definidos com uma variação de 2.5 a 3 desvios padrões. Para os nossos dados utilizarei definirei os outliers como aqueles que os valores variam mais que 3 vezes do desvio padrão. (FALAR O PORQUE)

- Além disso, existem apartamentos ou quartos com valor '0' o que provalvelmente é um erro, e dessa forma também serão retirados da análise.

```python
Dados_filtrado = Dados_treino[Dados_treino['price_z'] < 3] 
Dados_filtrado = Dados_filtrado[Dados_filtrado['night_z'] < 3]
Dados_filtrado = Dados_filtrado[Dados_filtrado['price'] != 0]
```

- Após aplicar os filtros é possível observar que houve um redução de um pouco menos de 300 valores
- Ainda existem valores que estão faltando 'números de reviews por mês' e 'data da última review recebi'. Assim,vamos tentar descobrir quais são esses valores que estão faltando para não perder essas variáveis

```python

# Achar os valores 'nan' da variável 'reviews_per_month' e depois ver quantos 'number _of_review' são zeros
finding_nan = Dados_filtrado[Dados_filtrado['reviews_per_month'].isnull()]
finding_nan['number_of_reviews'].value_counts()
```

- Pode-se observar que o valor de 'Nan' para 'reviews_per_month' são so mesmo que tem 'number _of_review' = 0. Dessa forma vamos substitur os valores 'nan' dos  'reviews_per_month' para zero.  

```python
Dados_filtrado['reviews_per_month'] = Dados_filtrado['reviews_per_month'].fillna(0)
Dados_filtrado.info()
```
